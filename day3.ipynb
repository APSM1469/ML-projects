# How do I apply filter criteria to a Panda dataframe?

```python
import pandas as pd
```

```python
movies = pd.read_csv('imdb.csv')
movies
```

```python
movies[movies['duration'] >=200]
```

```python
# Logical operators:
# AND OR
# AND - true only only if both sides of the operator are true (both the conditions should be true)
print(True and True)
```

```python
# OR - True if either side of the operator is true(atleast one of the condition should be true)
print(False or False)
```

```python
# 1. duration>=200
# 2. genre = drama
movies[(movies['duration']>=200) & (movies['genre'] =='Drama')]
```

# How do I handle missing values in Pandas

```python
# Based on business objective and dataset, analysts decides whethere to drop the null values
# or to impute values in place of null values
# 1. How to drop null values
```

```python
ufo = pd.read_csv('ufo.csv')
ufo
```

```python
# 'isnull' returns a dataframe of boolean value( True if missing)
ufo.isnull()
```

```python
# 'notnull' returns a dataframe of boolean value( True if not missing)
ufo.notnull()
```

```python
# Use the 'isnull' command to filter the dataframe
ufo[ufo['City'].isnull()]
```

```python
# Count the no.of missing values in each series
ufo.isnull().sum()
```

```python
# Examine no.of rows:
ufo.shape
```

```python
# if 'any' values are missing in a row, then drop that entire row
ufo.dropna(how = 'any').shape
```

```python
# if 'all' values are missing in a row, then drop that entire row
ufo.dropna(how = 'all').shape
```

```python
# if 'any' values are missing in a row( considering only city and shape reported) then drop  that entire row
ufo.dropna(subset = ['City', 'Shape Reported'],how = 'any').shape
```

```python
# if 'all' values are missing in a row( considering only city and shape reported) then drop  that entire row
ufo.dropna(subset = ['City', 'Shape Reported'],how = 'all').shape
```

```python
# 2. Imputing dummy values in place of null values
```

```python
ufo.head()
```

```python
# value_counts - gives the count of occurences of each unique value in a column
# By default, it does not include information related to missing values
ufo['Shape Reported'].value_counts(dropna = False) 
```
