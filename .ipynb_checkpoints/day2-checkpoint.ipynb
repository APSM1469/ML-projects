# How do I rename columns in a Pandas dataframe

```python
ufo = pd.read_csv("ufo.csv")
ufo
```

```python
ufo.columns
```

```python
# Rename two of the columns using 'rename' method:(temporarily)
# {key1:value1,key2:value2}
#[ele1,ele2..]
ufo.rename(columns = {'Colors Reported':'colors_reported','Shape Reported':'shape_reported'}, index = {0:'a'})
```

```python
ufo.head()
```

```python
# Rename two of the columns using 'rename' method:(permanently)
ufo.rename(columns = {'Colors Reported':'colors_reported','Shape Reported':'shape_reported'}, inplace = True)
```

```python
ufo.head()
```

```python
ufo.head()
```

```python
# Rename all of the column names by overwriting the 'columns' attribute
ufo_cols = ['city','colors reported','shape reported','state','time']
ufo.columns = ufo_cols
ufo.columns
```

```python
ufo.head()
```

# How can I drop multiple rows and columns at once

```python
ufo.head()
```

```python
# Remove multiple columns at a time:(temporarily)
ufo.drop(['city','colors reported'], axis = 1)
```

```python
ufo.head()
```

```python
# Remove multiple columns at a time:(permanently)
ufo.drop(['city','colors reported'], axis = 1, inplace = True)
```

```python
ufo.head()
```

```python
# Remove multiple rows at a time:(temporarily)
ufo.drop([0,1])
```

```python
# Remove multiple rows at a time:(permanently)
ufo.drop([0,1], inplace = True)
```

```python
ufo.head()
```

# How do I sort pandas Dataframe or series

```python
# By default, Python sorts in ascending order
```

```python
movies.head()
```

```python
# Sort the 'title' in ascending order( returns series of movie's titles in ascending order)
movies['title'].sort_values()
```

```python
# Sort the 'title' in descending order( returns series of movie's titles in descending order)
movies['title'].sort_values(ascending = False)
```

```python
# Sorting entire dataframe in ascending order by title
movies.sort_values('title')
```

```python
# Sorting entire dataframe in descending order by title
movies.sort_values('title',ascending = False)
```

```python
# Sorting entire dataframe in ascending order by content rating and duration
movies.sort_values(['content_rating','duration'])
```

```python
# Sorting entire dataframe in descending order by content rating and duration
movies.sort_values(['content_rating','duration'], ascending = False)
```
