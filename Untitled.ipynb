## How do I select pandas series from a dataframe

```python
# In Pandas, columns are referred as series
```

```python
ufo = pd.read_csv('ufo.csv')
ufo
```

```python
# There are two ways to access a series/column
# [ square bracket type] - most widely used
# syntax - dataframename[columname]
ufo['City']
```

```python
# dot notation type - cant be used if col name has white spaces
# syntax - dataframename.columname
ufo.City
```

```python
ufo.Colors Reported
```

```python
ufo['Colors Reported']
```
